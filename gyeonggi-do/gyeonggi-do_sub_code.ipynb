```python
import pandas as pd

df = pd.read_excel('/Users/home/Desktop/aa.xlsx')
```

```python
df = df.rename(columns= {'오전(%)':'평일 오전','낮(%)':'평일 낮','오후(%)':'평일 오후','오전(%).1':'주말 오전','낮(%).1':'주말 낮','오후(%).1':'주말 오후'})
a1 = df[df['행정구역']=='경기도'].index
df = df.drop(a1)
df['행정구역'] = df['행정구역'].str[1:4]
df
```

```python
import matplotlib.pyplot as plt
from matplotlib import rc
import seaborn as sns
%matplotlib inline

rc('font', family='AppleGothic')
plt.figure(figsize=(20,10))

x_values = df['행정구역']
y_values_0 = df['평일 오전']
y_values_1 = df['평일 오후']
plt.plot(x_values, y_values_0)
plt.plot(x_values, y_values_1)

plt.legend(['평일 오전','평일 오후'])

plt.show()
```
